```python
'''Esta aplicação entrará no sistema ERP, onde fará o download de todos os arquivos de 
transportes contidos na relação.
'''
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from time import sleep
from openpyxl import load_workbook

#Carrega a Planilha onde os dados que alimentarão o site estarão

wb = load_workbook(filename = 'enviarBoleto.xlsm')
sh = wb.active

linha = sh.cell(2 , 15)
row = 2
column = 1

servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)

#tela de Login / usuário / senha
navegador.get('http://192.168.0.10/WTGP751A2.aspx')

usuario = input('Por favor digite o seu usuário: ')
senha = input('Por favor digite sua senha: ')

navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_txtUserId').send_keys(usuario)

navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_txtSenha').send_keys(senha)
navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_btnOk').click()
sleep(1) 
navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_btnEntra').click()
sleep(2) 

#Nesta tela o programa irá recolher as informações contidas na planilha

for i in range(9, row + 1):
    filial = sh.cell(i , 4)
    cte = sh.cell(i, 5) 
    
    
    navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtCodigoFilial').send_keys(filial.value)
    sleep(1)
    
    navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtNrFaturaDe').send_keys(cte.value)
    sleep(0.5)
    navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtNrFaturaDe').send_keys(cte.value)
    sleep(0.5)
    navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtNrFaturaAte').send_keys(cte.value)
    sleep(0.5)

    
    navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_tabsFiltros_tabPeriodo_txtDataDe').send_keys("01112022")
    navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_tabsFiltros_tabPeriodo_txtDataAte').send_keys("31122022")
    sleep(0.5) 
   
   

   
    navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_btnBusca').click()
    sleep(0.5) 
    navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_gvConhecimentos_ctl02_imgDownloadAnexo').click()
    sleep(1)
                  
                  
    row = linha.value
    for i in range(3, row + 1):
        filial = sh.cell(i , 4)
        cte = sh.cell(i, 5)
        
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtNrFaturaAte').clear()
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtNrFaturaAte').send_keys(cte.value)
        sleep(0.2)
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtNrFaturaDe').click()
        sleep(0.2)
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtNrFaturaDe').clear()
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtNrFaturaDe').send_keys(cte.value)
        sleep(0.2)
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtCodigoFilial').click()
        sleep(0.2)
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtCodigoFilial').clear()
        sleep(0.2)
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_txtCodigoFilial').send_keys(filial.value)
        sleep(0.2)
        
      

        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_btnBusca').click()
        sleep(0.2) 
        navegador.find_element(By.ID, 'ctl00_ContentPlace_Main_tabConheciemntoCteNfe_TabPanelCte_gvConhecimentos_ctl02_imgDownloadAnexo').click()
        sleep(1)
```
